```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-2, 1)
x2 = np.linspace(1, 2)
x3 = np.linspace(2, 3)
x4 = np.linspace(3, 4)
x5 = np.linspace(4, 5)
x6 = np.linspace(5, 6)
x7 = np.linspace(6, 8)
 
plt.plot(x1, [0 for x in x1], 'b',  label='y = 0, para x < 1')
plt.plot(1, 0, 'ro', markersize=5, markerfacecolor='w')

plt.plot(x2, [1/6 for x in x2], 'g', label='y = 1/6, para x en [1, 2)')
plt.plot(1, 1/6, 'ro', markersize=5)
plt.plot(2, 1/6, 'ro', markersize=5, markerfacecolor='w')

plt.plot(x3, [2/6 for x in x3], 'r', label='y = 2/6, para x en [2, 3)')
plt.plot(2, 2/6, 'ro', markersize=5)
plt.plot(3, 2/6, 'ro', markersize=5, markerfacecolor='w')

plt.plot(x4, [3/6 for x in x4], 'y', label='y = 3/6, para x en [3, 4)')
plt.plot(3, 3/6, 'ro', markersize=5)
plt.plot(4, 3/6, 'ro', markersize=5, markerfacecolor='w')

plt.plot(x5, [4/6 for x in x5], 'gray', label='y = 4/6, para x en [4, 5)')
plt.plot(4, 4/6, 'ro', markersize=5)
plt.plot(5, 4/6, 'ro', markersize=5, markerfacecolor='w')

plt.plot(x6, [5/6 for x in x6], 'c', label='y = 5/6, para x en [5, 6)')
plt.plot(5, 5/6, 'ro', markersize=5)
plt.plot(6, 5/6, 'ro', markersize=5, markerfacecolor='w')

plt.plot(x7, [1 for x in x7], 'k', label='y = 1, para x > 6')
plt.plot(6, 1, 'ro', markersize=5)
plt.xlabel('x')
plt.ylabel('F(x)')
 
#plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           #ncol=2, mode="expand", borderaxespad=0.)
 
plt.grid()
plt.show()
```

```python
x_values = [0, 1, 2]
f_X_values = [1/4, 1/2, 1/4]

# Gráfico de barras
plt.bar(x_values, f_X_values, width=0.1, color='b', alpha=0.75)

# Etiquetar ejes
plt.xlabel('x')
plt.ylabel('f(x)')
# Establecer rango eje Y
plt.ylim(0, 0.6)


# Título
#plt.title('Función de densidad de la variable aleatoria discreta X')

for i, v in enumerate(f_X_values):
    plt.text(x_values[i] - 0.05, v + 0.02, str(v), color='black')

# Mostrar cuadrícula
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.show()
```
